```python
import pandas as pd

# Mapping function to group `trainer-gallery` and `shiny-vault` under their main set names
def group_set_names(set_name):
    if 'trainer-gallery' in set_name:
        return set_name.replace('-trainer-gallery', '')
    elif 'shiny-vault' in set_name:
        return set_name.replace('-shiny-vault', '')
    elif 'galarian-gallery' in set_name:
        return set_name.replace('-galarian-gallery', '')
    return set_name

# Apply the mapping function to create a new column
filtered_df['set_name_grouped'] = filtered_df['set_name'].apply(group_set_names)

# Display unique values in the new grouped column
grouped_set_names = filtered_df['set_name_grouped'].unique()
print(grouped_set_names)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Example rarity and printing to compare
selected_printing = 'Holofoil'
selected_rarity = 'Holo Rare'

# Filter data for the selected rarity and printing
filtered_data = filtered_no_pull_rate[
    (filtered_no_pull_rate['printing'] == selected_printing) &
    (filtered_no_pull_rate['rarity'] == selected_rarity)
]

# Group data by set name
sets = filtered_data['set_name_grouped'].unique()
grouped_data = {set_name: filtered_data[filtered_data['set_name_grouped'] == set_name]['price'] for set_name in sets}

# Perform pairwise Kolmogorov-Smirnov test for each set pair
ks_results = []
set_list = list(grouped_data.keys())
for i, set_a in enumerate(set_list):
    for set_b in set_list[i + 1:]:
        ks_stat, p_value = stats.ks_2samp(grouped_data[set_a], grouped_data[set_b])
        ks_results.append({'set_a': set_a, 'set_b': set_b, 'ks_stat': ks_stat, 'p_value': p_value})

# Convert results to a DataFrame and display
import pandas as pd
ks_df = pd.DataFrame(ks_results)
ks_df['significant'] = ks_df['p_value'] < 0.05  # Add a significance column
print(ks_df)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def compare_distributions_by_release_date(df, rarity, printing, save_plot=False, plot_path='distribution_plot.png'):
    """
    Compare price distributions for a specified rarity and printing with no pull rates,
    ordered by the release date from the oldest set to the newest.

    Parameters:
    - df (pd.DataFrame): DataFrame containing the card data with columns 'release_date', 'set_name_grouped', 'rarity', 'printing', and 'price'.
    - rarity (str): The rarity to analyze (e.g., 'Common', 'Uncommon').
    - printing (str): The printing to analyze (e.g., 'Normal', 'Holofoil').
    - save_plot (bool): Whether to save the generated plot to a file. Default is False.
    - plot_path (str): File path to save the plot if save_plot=True. Default is 'distribution_plot.png'.

    Returns:
    - None: Displays a boxplot comparing the distributions.
    """
    # Filter for the specified rarity and printing, with no pull rates
    filtered_data = df[
        (df['rarity'] == rarity) &
        (df['printing'] == printing) &
        (df['precise_pull'].isna())
    ].copy()
    
    # Ensure release_date is sorted chronologically
    filtered_data['release_date'] = pd.to_datetime(filtered_data['release_date'])
    filtered_data.sort_values(by='release_date', inplace=True)

    # Add a combined label for x-axis: "Date (Set Name)"
    filtered_data['date_label'] = filtered_data['set_name_grouped'] + ' - ' + filtered_data['release_date'].dt.strftime('%Y-%m-%d') 

    # Check if there is data to plot
    if filtered_data.empty:
        print("No data available for the specified rarity and printing.")
        return

    # Plot the distributions
    plt.figure(figsize=(14, 8))
    sns.boxplot(data=filtered_data, x='date_label', y='price', order=filtered_data['date_label'].drop_duplicates())
    plt.title(f'Price Distribution for {rarity} ({printing}) by Release Date')
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('Release Date (Set Name)')
    plt.ylabel('Price')
    plt.tight_layout()

    # Save the plot if requested
    if save_plot:
        plt.savefig(plot_path)
        print(f"Plot saved to {plot_path}")

    # Display the plot
    plt.show()
```

```python
def calculate_booster_pack_value(df):
    """
    Calculate the expected value of a booster pack for each set.

    Parameters:
    - df (pd.DataFrame): DataFrame containing card data with columns 'set_name_grouped', 'price', 'precise_pull', 'rarity'.

    Returns:
    - pd.DataFrame: DataFrame with columns ['set_name_grouped', 'expected_value'].
    """
    results = []

    # Group by set
    grouped = df.groupby('set_name_grouped')

    for set_name, group in grouped:
        # Step 1: Bulk cards (cards with NaN pull rates)
        bulk_cards = group[group['precise_pull'].isna()]
        bulk_price_avg = bulk_cards['price'].mean() if not bulk_cards.empty else 0
        bulk_value = bulk_price_avg * 10

        # Step 2: Cards with pull rates (including code cards)
        cards_with_pull_rates = group[group['precise_pull'].notna()]
        pull_rate_value = (cards_with_pull_rates['precise_pull'] * cards_with_pull_rates['price']).sum()

        # Step 3: Total expected value
        expected_value = bulk_value + pull_rate_value

        # Append result for this set
        results.append({'set_name_grouped': set_name, 'expected_value': expected_value})

    # Convert results to a DataFrame
    return pd.DataFrame(results)
```
